```python
with open("digits.bin", "rb") as f:
    content = f.read()
print(content[:30])
```

```python
assembled_data = b""
for i in range(len(content) // 8):
    chunk = content[8*i: 8*i+8]

    assembled_byte = 0
    for j in range(len(chunk)):
        if chunk[j] == ord('1'):
            assembled_byte += (1 << (7-j))
    
    assembled_data += assembled_byte.to_bytes(1)
print(assembled_data[:30])
    
```

JFIF suggests that this is a JPG image

```python
with open("assembled.jpg", "wb") as f:
    f.write(assembled_data)
# You should now see the flag in assembled.jpg
```
