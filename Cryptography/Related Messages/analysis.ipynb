```python
#p1.quo_rem(p2)

def gcd_franklin_reiter(a, b):
    while b:
        q, r = a.quo_rem(b)
        a, b = b, r
    return a.monic()

# If gcd_franklin_reiter succeeds, the constant term is -m_1, so recover message.
m1 = int((-gcd_franklin_reiter(p1, p2).list()[0])) % N
```
