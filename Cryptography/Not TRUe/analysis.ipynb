# Goal
Find $f$ and $g$ from $h \in \mathbb{Z}_q[x]/(x^N-1)$

Note that $h = p f_q^{-1} g \mod q$, so
$$
h f = p g \mod q
$$
which is a linear system

## Attack Plan
Let $f, g \in \mathbb{Z}_q[x]$ and since cyclic convolution of $x^N-1$ (so that $x^N \equiv 1 \mod q$)

$$
\begin{pmatrix}
f_0 & f_1 & \cdots & f_{N-1}
\end{pmatrix}
\begin{pmatrix}
h_0 & h_1 & \cdots & h_{N-1} \\
h_{N-1} & h_0 & \cdots & h_{N-2} \\
h_{N-2} & h_{N-1} & \cdots & h_{N-3} \\
\vdots & \vdots & \ddots & \vdots\\
h_1 & h_2 & \cdots & h_0
\end{pmatrix}
=
\begin{pmatrix}
p g_0 & p g_1 & \cdots & p g_{N-1}
\end{pmatrix}
$$
which can be rewritten as 
$$
\begin{pmatrix}
f_0 & f_1 & \cdots & f_{N-1} & g_0 & g_1 & \cdots & g_{N-1}
\end{pmatrix}
\begin{pmatrix}
h_0 & h_1 & \cdots & h_{N-1} \\
h_{N-1} & h_0 & \cdots & h_{N-2} \\
h_{N-2} & h_{N-1} & \cdots & h_{N-3} \\
\vdots & \vdots & \ddots & \vdots\\
h_1 & h_2 & \cdots & h_0 \\
-p & 0 & \cdots & 0 \\
0 & -p & \cdots & 0 \\
\vdots & \vdots & \ddots & 0 \\
0 & 0 & \cdots & -p
\end{pmatrix}
=
0
\mod q
$$

Taking modulo $q$ into account
$$
\begin{pmatrix}
f_0 & f_1 & \cdots & f_{N-1} & g_0 & g_1 & \cdots & g_{N-1} & k_0 & k_1 \cdots & k_{N-1}
\end{pmatrix}
\begin{pmatrix}
h_0 & h_1 & \cdots & h_{N-1} \\
h_{N-1} & h_0 & \cdots & h_{N-2} \\
h_{N-2} & h_{N-1} & \cdots & h_{N-3} \\
\vdots & \vdots & \ddots & \vdots\\
h_1 & h_2 & \cdots & h_0 \\
-p & 0 & \cdots & 0 \\
0 & -p & \cdots & 0 \\
\vdots & \vdots & \ddots & 0 \\
0 & 0 & \cdots & p \\
q & 0 & \cdots & 0 \\
0 & q & \cdots & 0 \\
\vdots & \vdots & \ddots & 0 \\
0 & 0 & \cdots & q
\end{pmatrix}
= 0
$$

```python
# Use optimized version of 2N x 2N
# I H
# 0 qI
H = matrix(ZZ, N, N, [h[N-i:] + h[0:N-i] for i in range(N)])
B = matrix(ZZ, 2*N, 2*N, 0)
B[0:N,N:] = H
B[0:N, 0:N] = matrix.identity(N)
B[N:, N:] = matrix.identity(N) * q
```

```python
B_reduced = B.LLL()
```

```python
def decrypt(f, ciphertext, p, q, R_modp, R_modq):
    # 1. Compute f * c in R_modq
    a = R_modq(f) * ciphertext
    
    # 2. Center Lift: Move coefficients to [-q/2, q/2]
    # This is vital for the modular reduction to stay valid over integers
    a_coeffs = []
    for coeff in a.list():
        c = Integer(coeff)
        if c > q // 2:
            c -= q
        a_coeffs.append(c)
    
    # 3. Move to R_modp and multiply by f_p_inv
    # Note: f_p_inv must be calculated in R_modp
    f_p_inv = R_modp(f)**-1
    m_poly = R_modp(a_coeffs) * f_p_inv
    
    return m_poly

# Use it on your ciphertexts
recovered_ms = [decrypt(f_recovered, R(c), p, q, R_modp, R_modq) for c in ct]
```

```python
def polys_to_text(polys, N):
    bit_str = ""
    for p in polys:
        # Get coefficients, pad with zeros to length N
        coeffs = p.list()
        coeffs += [0] * (N - len(coeffs))
        bit_str += "".join(str(int(c)) for c in coeffs)
    
    # Convert bits to bytes (8 bits at a time)
    chars = []
    for i in range(0, len(bit_str), 8):
        byte = bit_str[i:i+8]
        if len(byte) < 8: break
        chars.append(chr(int(byte, 2)))
        
    return "".join(chars)

# Redacted
#print("Flag:", polys_to_text(recovered_ms, N))
```
