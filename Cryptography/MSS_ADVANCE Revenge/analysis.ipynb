```python
A_list = []
for x, _ in pairs:
    row = []
    for i in range(30):
        row.append(pow(x, 29-i, p))
    A_list.append(row)

A = matrix(ZZ, 20, 30, A_list)
A = A.transpose()
A
```

```python
B = matrix(ZZ, 51, 51, 0)
for i in range(20):
    #B[i, i] = 1
    B[i, i] = p

B[20:50, :20] = A

for i in range(len(pairs)):
    y = pairs[i][1]
    B[50, i] = -y

B[20:50, 20:50] = matrix.identity(30)

# Weight
B[50, 50] = p
```

```python
B_reduced = B.LLL()
```

```python
for row in B_reduced:
    # We want the row where the relation was satisfied.
    # This row will usually have the weight (p) or -p at index 50.
    if abs(row[50]) == p:
        # The coefficients are in indices 20 to 49
        # Remember: if row[50] is -p, you might need to flip the signs of c_i
        coeffs = row[20:50]
        if row[50] == -p:
            coeffs = [-c for c in coeffs]
        
        print("Found coefficients: " + str(coeffs))
        break
```
